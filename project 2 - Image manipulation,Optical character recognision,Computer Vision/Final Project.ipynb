```python
import zipfile

from PIL import Image,ImageDraw
import pytesseract
import cv2 as cv
import numpy as np
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# loading the face detection classifier
face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

# the rest is up to you!
```

```python
def zip_process(path):
    """Takes the path of the zip file and returns a tuple with each tuple 
        having 
        1)file_name
        2)image file
        3)list of rectangle boxes with each giving information about each face detected
        4)the complete text detected in that image"""
    
    archive = zipfile.ZipFile(path)
    my_zip_tuples = []
    for entry in archive.infolist():
#         print('loop executed')
        file = archive.open(entry)
        img = Image.open(file)
        opencvImage = cv.cvtColor(np.array(img), cv.COLOR_RGB2BGR)
        faces = face_cascade.detectMultiScale(opencvImage,1.35,minNeighbors = 4)
        try:
            faces = faces.tolist()
            text = pytesseract.image_to_string(img)
            my_zip_tuples.append((file.name,img.copy(),faces,text))
        except:
            np.array(faces).tolist()
            text = pytesseract.image_to_string(img)
            my_zip_tuples.append((file.name,img.copy(),faces,text))
    return my_zip_tuples


def get_faces_from_articles_having_required_string(path_of_zip_file,string_to_search):
    """takes in path of the zip file and returns the faces from all the files 
        in that zip file that have the given string in that file"""
         
    path = path_of_zip_file
    tuples = zip_process(path)
    for a,b,c,d in tuples:
        if string_to_search in d:
            images = []
            print("Results found in file {}".format(a))
            for i in c:
                box_to_crop = (i[0],i[1],i[0]+i[2],i[1]+i[3])
                img = b.crop(box_to_crop)
                size = 128,128
                img.thumbnail(size)
                images.append(img)
            if len(images) < 1:
                print("But there were no faces in that file!")
            else:
                sorted_images_according_to_sizes=sorted(images,key = lambda x:x.size[0],reverse = True)
                first_image = sorted_images_according_to_sizes[0] 
                if len(images)>5:
                    contact_sheet=Image.new(first_image.mode, (first_image.width*5,2*first_image.height))
                else:
                    contact_sheet=Image.new(first_image.mode, (first_image.width*5,first_image.height))
                x=0
                y=0
                for img in sorted_images_according_to_sizes:
                    contact_sheet.paste(img, (x, y) )
                    if x+first_image.width == contact_sheet.width:
                        x=0
                        y=y+first_image.height
                    else:
                        x=x+first_image.width
                display(contact_sheet)
```
